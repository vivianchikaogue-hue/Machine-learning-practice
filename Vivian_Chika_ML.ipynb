# **Import Libraries and Data**

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn import linear_model
from google.colab import drive
drive.mount('/content/drive')
df=pd.read_csv('/content/drive/MyDrive/hiring.csv')
```

# View Dataset

```python
df
```

# Cleaning the Dataset




```python
#finding the median test score

median_test_score = df['test_score(out of 10)'].median()
median_test_score
```

```python
#Filling in the missing test score with the median value

df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score)
df['test_score(out of 10)']
```

```python
#Filling the missing values in the experience column and converting the words to numbers

df['experience']  = df['experience'] .fillna("zero")

!pip install word2number
from word2number import w2n
df['experience'] = df['experience'].apply(w2n.word_to_num)
df['experience']
```

# **Model Training**

```python
# Defining our axes
y=df['salary($)']
x=df.drop('salary($)',axis=1)

#training our model
reg=linear_model.LinearRegression()
reg.fit(x,y)
```

# **Making predictions**

```python
# predict salaries for following candidates, 2 yr experience, 9 test score, 6 interview score,
# and 12 yr experience, 10 test score, 10 interview score
Candidate_A = reg.predict([[2,9,6]])
Candidate_B = reg.predict([[12,10,10]])
print(Candidate_A)
print(Candidate_B)
```

```python
#The predicted salaries for both candidates are
Salary_for_Candidate_A = round(Candidate_A[0])
Salary_for_Candidate_B = round(Candidate_B[0])
print(Salary_for_Candidate_A)
print(Salary_for_Candidate_B)
```

```python
# Returns the R^2 score
score = reg.score(x, y)
print(score)
```
